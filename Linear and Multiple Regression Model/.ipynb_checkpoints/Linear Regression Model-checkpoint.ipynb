```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('five_minute.csv', parse_dates= ['time'], index_col= 'time')
```

```python
data.info()
```

```python
data
```

```python
data.plot(figsize = (12,8))
plt.show()
```

```python
data.loc['2019-06'].plot(figsize = (12,8))
plt.show()
```

```python
data['returns'] = np.log(data.div(data.shift(1)))
```

```python
data
```

# A Simple Linear Model to predict the financial returns

```python
data['lag1'] = data.returns.shift(1) #returns shifted by 1
```

```python
data.dropna(inplace= True)
```

```python
data
```

```python
data.iloc[:, -2:].plot(kind = 'scatter', x ='lag1', y = 'returns')
plt.xlim(-0.005, 0.005)
plt.ylim(-0.005, 0.005)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression(fit_intercept= True)
```

### Independent Variable(s) = lag1
### Dependent Variable = Returns

```python
lm.fit(data.lag1.to_frame(), data.returns)
```

```python
slope = lm.coef_
slope
```

```python
intercept = lm.intercept_
intercept
```

```python
data['predictions'] = lm.predict(data.lag1.to_frame())
data
```

```python
plt.figure(figsize = (12,8))
plt.scatter(x = data.lag1, y = data.returns, label = 'Data')
plt.plot(data.lag1, data.predictions, c= 'red', label = 'Linear Regression')
plt.xlim(-0.005, 0.005)
plt.ylim(-0.005, 0.005)
plt.legend(fontsize = 12)
plt.xlabel('Lag1 returns', fontsize = 13)
plt.ylabel('Returns', fontsize = 13)
plt.show()
```

```python
data[['returns', 'predictions']].plot(figsize = (12,8))
plt.show()
```

# This model fails to predict the magnitude of returns

```python
data.predictions = np.sign(data.predictions) 
# Trying to find out if the data can predict the market directions(+1/-1)
```

```python
data
```

```python
np.sign(data.predictions * data.returns)
```

```python
hits = np.sign(data.predictions * data.returns).value_counts()
hits
```

```python
hit_ratio = hits[1.0]/sum(hits)
hit_ratio # percentage of correctly predicted market returns
```
