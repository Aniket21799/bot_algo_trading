```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## SMA(Simple Moving Average) is a strategy generally used to generate buy and sell signals in the financial instruments using the simple moving averages. Most generally used pair for SMA is 50 and 200. Buy signal is generated when the SMA 50 crosses the SMA 200 from bottom and a sell signal is generated when SMA 50 crosses the SMA 200 from above.

```python
data = pd.read_csv('eurusd.csv', parse_dates = ['Date'], index_col = 'Date')
data
```

```python
sma_s = 50
sma_l = 200
# Defining short and long moving averages
```

```python
data['SMA_S'] = data.price.rolling(sma_s).mean()
data['SMA_L'] = data.price.rolling(sma_l).mean()
#rolling takes a series of data, a window and a function to apply to the window.
```

```python
data
```

```python
data.dropna(inplace= True)
```

```python
data.plot(figsize = (12,8), title = 'EUR/USD- SMA{} | SMA{}'.format(sma_s, sma_l), fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
```

```python
# SMA_S and SMA_L for the year 2016
data.loc['2016'].plot(figsize = (12,8), title = 'EUR/USD- SMA{} | SMA{}'.format(sma_s, sma_l), fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
```

```python
data['position'] = np.where(data['SMA_S'] > data['SMA_L'], 1, -1)
data
# position that should be taken according to SMA
```

```python
data.loc[:, ['SMA_S', 'SMA_L', 'position']].plot(figsize = (12,8), title = 'EUR/USD- SMA{} | SMA{}'.format(sma_s, sma_l), fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
# The values will range from 0 to 1 depicting buy or sell positions.
```

# Vectorized Strategy Backtesting

```python
data['returns'] = np.log(data.price.div(data.price.shift(1)))
```

```python
data['r_strategy'] = data.position.shift(1) * data['returns']
data
# The postion column depicted the positin for the NEXT DAY. Hence r_strategy shifted the position by one and multiplied by the
# returns(returns of buy and hold)
```

```python
data.dropna(inplace= True)
```

```python
data[['returns', 'r_strategy']].sum() # absolute performance
```

```python
data[['returns', 'r_strategy']].sum().apply(np.exp)
```

```python
data[['returns', 'r_strategy']].mean() * 252 # annualized returns
```

```python
data[['returns', 'r_strategy']].std() * np.sqrt(252) # annualized risk
```

```python
data['cumreturns'] = data['returns'].cumsum().apply(np.exp)
data['cumstrategy'] = data['r_strategy'].cumsum().apply(np.exp)
```

```python
data
```

```python
data[['cumreturns', 'cumstrategy']].plot(figsize = (12,8), title = 'EUR/USD- SMA{} | SMA{}'.format(sma_s, sma_l), fontsize = 12)
plt.legend(fontsize = 12)
plt.show()
# performance comparison between buy and hold and SMA strategy
```

```python
outperformace = data.cumstrategy.iloc[-1] - data.cumreturns.iloc[-1]
outperformace
```
